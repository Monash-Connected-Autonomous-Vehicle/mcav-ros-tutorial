---

The primary differences between Python and C++ are:
- Higher Level: C++ allows and often requires lower level operations such as addressing memory locations. Python abstract most of these lower level operations away from the user, and handles it automatically. This can make it easier to write, but gives less control over how things are done. 
- Interpreted: Python does not get compiled before being run. The computer interprets the code line by line as it is executed. While this lets you skip the compile stage, your computer often won't pick up errors until the line in question is executed. These are called runtime errors. 
- Types: Python code has types, but they are not strongly enforced. This means you can change variables on the fly, and the interpretter won't complain. However, because they aren't enforced you can accidently change a variable and no error will occur which can lead to difficult to find bugs. 
- Readability: Python is designed to be as close to human readable as possible. Python uses words for many operations which makes it very easy to understand, even if you haven't seen it before. 

For these reasons, Python is a lot easier to write and understand than C++, but much, much slower. We mainly use Python for software that doesn't need to run in real-time (e.g. not on the car). For these use cases Python is a great choice.

#### Question 3.1 Solution
The following cell contains an example solution for Q3.1. Your solution might look different. 

```python
#@title #### Question 3.1 - Example Solution
%matplotlib inline

# Initial Setup
n_timesteps = 100

# Controller Class
class Controller():
    
    # initialisation
    def __init__(self):
        self.position = 0
        self.action = 0
        self.difference = 0
        self.position_history = []
        self.action_history = []

    def calculate_difference(self, target):
        # Calculate the difference between the target and current position
        self.difference = target - self.position

    def calculate_action(self):
        # Calculate the movement action requierd (max distance of 2)
        self.action = np.clip(self.difference, -2, 2)

    def move_position(self):
        # Update controller position based on action calculated
        self.position += self.action

    def add_history(self):
        # Append the current position and action to history list
        self.position_history.append(self.position)
        self.action_history.append(self.action)
        

# Instantiate controller
controller = Controller()

# Calculate Target
targets = np.full(shape=n_timesteps, fill_value=10)


# Loop through all target values
for i, target in enumerate(targets):

    # Calculate difference between target and current position
    controller.calculate_difference(target)

    # Calculate required action (max distance of 2)
    controller.calculate_action()

    # Update controller position
    controller.move_position()

    # Randomly perturb the position of the controller
    perturbation = np.random.randint(-3, 3)
    controller.position += perturbation

    # Append current position and action to history
    controller.add_history()


# Plotting Target, Position, and Action History
plt.figure(figsize=(10, 10))
plt.title("Line Follower Controller")
plt.plot(targets, label="target")
plt.plot(controller.position_history, label="position")
plt.plot(controller.action_history, label="action")
plt.legend()
plt.grid("on")
plt.show()

```
